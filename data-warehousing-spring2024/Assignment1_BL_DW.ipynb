```python
from statistics import mean
import csv
import re
import pandas as pd
```

```python
# declaring list
commaslist = []

with open("StudentsPerformance-1.csv",'r') as StuPerf:
    # convert to upper case
    # find and replace all non alphanumeric characters with regex
    # split lines with commas and append to commaslist
    for line in StuPerf:
        line = line.upper()
        line = re.sub(r'[^a-zA-Z0-9,+]+', '',line)
        comma = line.split(',')
        
        #append each line to list
        commaslist.append(comma)
```

```python
#print list
commaslist
```

```python
# not a function, so it's a little long
mathscorelist = []
readingscorelist = []
writingscorelist = []

# getting all math, reading, and writing scores and putting into list
for line in commaslist:
    mathscore = line[5]
    mathscorelist.append(mathscore)
    
    readingscore = line[6]
    readingscorelist.append(readingscore)
    
    writingscore = line[7]
    writingscorelist.append(writingscore)
    
# popping off header and converting each string to int
mathscorelist.pop(0)
mathscorelist = list(map(int, mathscorelist))

readingscorelist.pop(0)
readingscorelist = list(map(int, readingscorelist))

writingscorelist.pop(0)
writingscorelist = list(map(int, writingscorelist))

# using statistics package, print mean of each category of scores
average_math = mean(mathscorelist)
print(average_math)

average_reading = mean(readingscorelist)
print(average_reading)

average_writing = mean(writingscorelist)
print(average_writing)
```

```python
# saving list of new upper case and processed strings to a dataframe
df = pd.DataFrame(commaslist)
```

```python
# writing dataframe to new csv file
df.to_csv('StudentsPerformance_updated.csv', index=False, header=False)
```
